# The Weather Dataset

Here, the Weather Dataset is a time-series data set with per-hour information about the weather conditions at a particular location. It records Temperature, Dew Point Temperature, Relative Humidity, Wind Speed, Visibility, Pressure and Conditions.

This data is available as CSV file. We are going to analyze this data set using the Pandas DataFrame.

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'Weather Data.csv')
```

```python
data
```

# How to Analyze DataFrames ?

# .head()

It shows the first N rows in the data(by default, N=5).

```python
data.head()
```

# .shape

It shows the total number of rows and columns of the dataframe

```python
data.shape
```

# .index

This attribute provides the index of the dataframe

```python
data.index
```

# .columns

It shows the name of each column

```python
data.columns
```

# .dtypes

It shows the data-type of each column

```python
data.dtypes
```

# .unique()

In a column, it shows all the unique values present in the column. It can be applied on a single column only, not on the
whole dataframe.

```python
data['Weather'].unique()
```

# .nunique()

It shows the total no. of unique values in each column. It can be applied on a single column as well as on whole dataframe.

```python
data.nunique()
```

# .count()

It shows the total no. of non_null values in each column. It can be applied on single column as well as on the whole
dataframe.

```python
data.count()
```

# .value_counts()

In a column, it shows all the unique values with their count. It can be applied on single column only.

```python
data['Weather'].value_counts()
```

# .info()

Provides basic information about the dataframe.

```python
data.info()
```
